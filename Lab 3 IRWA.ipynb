```python
from nltk.tokenize import word_tokenize
#for ngram

from nltk.util import ngrams


text = "april is the month"
tokenze_terms = word_tokenize(text)
print(tokenze_terms)

#ngram this in token wise
print('ngram this in token wise\n')
n_grams=ngrams(tokenze_terms,3)
#print(list(n_grams))
x=[' '.join(n) for n in n_grams]
print(x)


#ngram this in character wise 
print('\n ngram this in character wise\n')
text = "april is the month"
#text to list

char_list=[]
for char in text:
  char_list +=char
print(char_list)
n_grams=ngrams(char_list,3)
x=[' '.join(n) for n in n_grams]
print(list(x))


```

**b).Write code to find out the Jaccard coefficient between the given documents based on different k**.

```python



def jaccard(x,y):
  numerator=x.intersection(y)
  intersect_ngram_count=len(numerator)
  denomenator=x.union(y)
  union_ngram_count=len(denomenator)
  return  intersect_ngram_count/union_ngram_count

doc3=open(os.getcwd()+'/ngram/doc3.txt','r')
#print(doc3.read())

doc4=open(os.getcwd()+'/ngram/doc4.txt','r')
#print(doc4.read())



doc_3_list=[char for char in doc3.read()]
#print(doc_3_list)

n_grams=ngrams(doc_3_list,2)
x=[' '.join(n) for n in n_grams]
print(x)

doc_4_list=[char for char in doc4.read()]
#print(doc_4_list)

n_grams=ngrams(doc_4_list,2)
y=[' '.join(n) for n in n_grams]
print(y)

print(jaccard(set(x),set(y)))
```
